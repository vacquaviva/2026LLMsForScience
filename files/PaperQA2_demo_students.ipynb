```python
llm="gpt-4.1-nano"

settings = Settings(
    llm=llm, 
    summary_llm=llm,
    #llm_config={"rate_limit": {"gpt-4o-mini": "30000 per 1 minute"}},
    #summary_llm_config={"rate_limit": {"gpt-4o-mini": "30000 per 1 minute"}},
    agent = AgentSettings(
        agent_llm=llm,
        #agent_llm_config={
        #    "rate_limit": {llm: "30000 per 1 minute"},
        #},
        index=IndexSettings(paper_directory="sources/"), #put your relevant folder
    ),
     parsing=ParsingSettings(enrichment_llm=llm,disable_doc_valid_check=True, use_doc_details=False),
)
```

```python
index = await get_directory_index(settings=settings, build=True)
```

```python
# Check files in index
print(await index.index_files)
```
